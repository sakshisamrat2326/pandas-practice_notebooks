
<h5 style="color:#2c3e50; font-size:48px;">ðŸ“Š 1)Finding Missing Data</h5>


```python
import pandas as pd
import numpy as np

# Sample dataset with np.nan
data = {
    "Name": ["Sakshi", "Aman", "Riya", "Karan", "Neha"],
    "Age": [23, np.nan, 21, 24, np.nan],
    "Department": ["IT", "HR", np.nan, "Finance", "IT"],
    "Salary": [50000, 45000, np.nan, 60000, 52000]
}

df = pd.DataFrame(data)
df
```

```python
df.isna() #Nan value=true , other given values=false
```

```python
df.isna().sum() #count of Na values in each row
```

```python
df.isna().any() # => any row that has no nan value is false
```
